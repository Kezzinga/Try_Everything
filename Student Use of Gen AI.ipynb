### Do college students believe that the use of generative AI could negatively impact their cognitive function?


##### Popular opinion would suggest that artificial intelligence is enhancing the lives of those who utilize it. People of many different professions are utilizing artificial intelligence in their careers and daily lives. Even students find their work load easier to manage using generative artifical intelligence tools such as ChatGPT. But what are the future implications? 

##### Young adults are the immediate future and their foundational knowledge developed in academia propels their futures. The use of generative AI in colleges and universities has increased immensley in the last few years. Though there are many advantages, there are also some drawbacks which have been identified in the responses to a questionnaire provided to Bangladeshi students. 

##### The students gave mostly positive reviews regarding the usefulness of Gen AI; however, many also believe that these tools can negatively impact their ability to critically think and foster creativity. A dataset containing responses of these students to ChatGPT has been analyzed below.

```python
# Load the dataset
impact = pd.read_excel('The Impact of AI and ChatGPT on Bangladeshi Students.xls')

# Clean up column names (remove leading/trailing spaces)
impact.columns = [col.strip() for col in impact.columns]

# Set the style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Helper function to create consistent count plots
def count_plot(column, title):
    plt.figure()
    ax = sns.countplot(
        data=impact,
        y=column,
        order=impact[column].value_counts().index,
        palette="viridis"
    )
    ax.set_title(title)
    ax.set_xlabel("Count")
    ax.set_ylabel("")
    plt.tight_layout()
    plt.show()

# List of visualizations to generate
count_plot("How often do you use AI tools like ChatGPT  for academic purposes?",
           "Frequency of AI Tool Use for Academic Purposes")

count_plot("Do you use ChatGPT for group study or collaborative learning ?",
           "Use of ChatGPT for Group Study")

count_plot("Do you feel that AI assists you in self-paced learning ?",
           "AI Assistance in Self-Paced Learning")

count_plot("Does ChatGPT help improve critical thinking and problem-solving?",
           "ChatGPT and Critical Thinking Improvement")

count_plot("Has ChatGPT provided useful guidance on your career path or educational choices ?",
           "ChatGPT's Role in Career or Educational Guidance")

count_plot("Do you think relying too much on AI tools like ChatGPT could harm students creativity and critical thinking skills?",
           "Perceived Risk of Overreliance on AI Tools")

count_plot("Do you believe using AI tools like ChatGPT reduces your creativity and critical thinking?",
           "Self-Perception: Does AI Reduce Creativity?")

count_plot("Do you believe AI tools (like ChatGPT) improve your learning experience?",
           "Belief That AI Tools Improve Learning Experience")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set plot style
sns.set(style="whitegrid")

# Bar plot showing the count of each response
plt.figure(figsize=(7, 5))
sns.countplot(data=impact, y="Do you believe using AI tools like ChatGPT reduces your creativity and critical thinking?", order=["Yes", "Occasionally", "No"])
plt.title("Student Beliefs on ChatGPT Reducing Creativity & Critical Thinking")
plt.xlabel("Number of Students")
plt.ylabel("Response")
plt.tight_layout()
plt.show()
```
